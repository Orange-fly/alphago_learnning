```python
    def showFrame( data):
        """Draw the board and show game info"""
        state = data[0]
        win   =data[2]
        height,width = state[0].shape         
        
        print("win is %d"  % win )
        print()
        for x in range(width):
            print("{0:6}".format(x), end='')
        print('\r\n')
        for i in range(height - 1, -1, -1):
            print("{0:3d}".format(i), end='')
            for j in range(width):
                player1 = state[0][i][j]
                player2 = state[1][i][j]
                if 1 == player1:
                    print('X'.center(6), end='')
                elif 1 == player2:
                    print('O'.center(6), end='')
                else:
                    print('_'.center(6), end='')
            print('\r\n\r\n')
```
